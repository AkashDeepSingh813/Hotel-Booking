### Data Analysis and Visualizations

```python
cancelled_perc=df['is_canceled'].value_counts(normalize='True')
cancelled_perc

plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
colors=['darkblue','orange']
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor='k',width=0.7,color=colors)
plt.show()
```

```python
# Define the custom colors for the countplot
colors = ['darkblue', 'orange']

plt.figure(figsize=(8, 4))

# Create the countplot with the custom palette
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette=colors)

# Set the custom legend labels
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(['Not Canceled', 'Canceled'], loc='upper left', bbox_to_anchor=(1, 1))

plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.tight_layout()  # Ensure everything fits within the figure area
plt.show()
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Use the same color scheme
resort_color = 'darkblue'
city_color = 'orange'

plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)

# Plot the data with the specified colors
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color=resort_color)
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', color=city_color)

plt.legend(fontsize=20)
plt.show()
```

```python
# Define the custom colors
colors = ['darkblue', 'orange']

# Add the 'month' column if not already added
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16, 8))

# Create the countplot with the custom color palette
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette=colors)

plt.title('Reservation Status by Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'], loc='upper right')
plt.tight_layout()  # Adjust the layout to fit everything properly
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)

# Specify 'x' and 'y' explicitly
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
# Calculate the average ADR for canceled reservations
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Calculate the average ADR for non-canceled reservations
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plotting the ADR for canceled and non-canceled reservations
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')

# Use the same color scheme as before
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Canceled', color='darkblue')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled', color='orange')

plt.legend()
plt.show()
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not_cancelled',color='darkblue')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled',color='orange')
plt.legend(fontsize=20)
plt.show()
```
